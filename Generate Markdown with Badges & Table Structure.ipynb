```python
def formatter(li : str) -> str:
    mon, tue, wed, thu, fri, sat, sun = li[0:2], li[2:5], li[5:8], li[8:11], li[11:14], li[14:17], li[17:20]
    formatted_dates = [v.strip().replace(v.strip(), v.strip().zfill(2)) for v in [mon, tue, wed, thu, fri, sat, sun]]
    
    # embed url string in markdown format
    uri_gen = "[![{date}](https://raw.githubusercontent.com/ZenithClown/markdown-calender/master/static/images/{date}.svg)](#)"
    formatted_dates = [uri_gen.format(date = d) if d != "00" else " " for d in formatted_dates]
    return " | ".join(formatted_dates)
```

```python
final_formatted_string = """
<td>
<div align = "center">

**`{title}`**
| MON | TUE | WED | THU | FRI | SAT | SUN |
| :-: | :-: | :-: | :-: | :-: | :-: | :-: |
{dates}

</div>
</td>
"""
```

```python
def total_generator(year, month):
    calendar_ = calendar.month(year, month).split("\n")
    
    # get paramateres from the generated calendar
    title = calendar_[0].strip().upper()
    dates = "\n".join(["| " + formatter(row) + " |" for row in calendar_[2:] if row])
    
    return final_formatted_string.format(title = title, dates = dates)
```

```python
print(total_generator(2022, 1))
```
